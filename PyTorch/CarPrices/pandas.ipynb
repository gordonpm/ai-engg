```python
import pandas as pd

df = pd.read_csv("../data/used_cars.csv")
df.head()
```

```python
df["model_year"].hist()
```

```python
df["model_year"].hist(bins=[2000,2010,2020,2030])
```

```python
print(df["model_year"].mean())
print(df["model_year"].median())
print(df["model_year"].min())
print(df["model_year"].max())
```

```python
price = df["price"]
price = price.str.replace("$","")
price = price.str.replace(",","")
price = price.astype(int)
print(price)
print(price.mean())
print(price.median())
```

```python
price[price < 200000].hist()
```

```python
df["price"] = price
df.head()
```

```python
df_hyundai = df[df["brand"] == "Hyundai"].head()
print(df_hyundai["price"].mean())

df_ford = df[df["brand"] == "Ford"].head()
print(df_ford["price"].mean())
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

plt.scatter(df["model_year"], df["price"])
plt.show()
```
